```sas
ods noproctitle;

proc sort data=SASHELP.CARS out=cars;
  by descending MSRP;
run;

proc print data=cars(obs=20);
  var Make Model Type;
  title "New autos";
run;
```

```sas
proc means data=cars chartype mean std min max n vardef=df;
  var MSRP MPG_Highway EngineSize Weight Horsepower;
  title "Summar statistics from cars data set";
  run;
```

```sas
proc univariate data=SASHELP.CARS vardef=df noprint;
  var MSRP MPG_Highway EngineSize;
  histogram MSRP MPG_Highway EngineSize / normal(noprint) kernel;
  inset mean std min max n / position=ne;
  title 'Distributions of Parameters from Cars Library';
run;
```
